## Matplotlib workflow examples

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [11, 22, 33 ,44]

fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(x, y)

ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

fig.savefig("simple-plot.png");
```

## Matplotlib with Numpy

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
## Line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
## Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
## Bar Charts
car_prices = {"Tesla model Y": 44000, "Toyota RAV4": 34000, "Ford MACHE": 40000, "Rivian": 50000}

fig, ax = plt.subplots()
ax.bar(car_prices.keys(), car_prices.values())
ax.set(title="Car Prices Plot", xlabel="Car models", ylabel="Car Prices");
```

```python
fug, ax = plt.subplots()
ax.barh(list(car_prices.keys()), list(car_prices.values()));
```

```python
## Histogram

x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Multiple options in subplots

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(car_prices.keys(), car_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(car_prices.keys(), car_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```
